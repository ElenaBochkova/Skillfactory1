```python
ratings_movies_df = pd.read_csv('data/ratings_movies.csv', sep=',')
display(ratings_movies_df)
```

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
ratings_movies_df['Movie year'] = ratings_movies_df['title'].apply(get_year_release)
display(ratings_movies_df)
```

```python
ratings_movies_df.isnull().sum()
```

```python
movie_df = ratings_movies_df[ratings_movies_df['Movie year'] == 1999].groupby(by = 'movieId').mean(numeric_only=True).sort_values('rating')

display(ratings_movies_df[ratings_movies_df['movieId'] == movie_df.index[:][0]])
```

```python
movie_df = ratings_movies_df[ratings_movies_df['Movie year'] == 2010].groupby(by = 'genres').mean(ratings_movies_df['rating'].all).sort_values('rating')
```

```python
display(movie_df)
```

```python
user_df = ratings_movies_df.groupby(by='userId')['genres'].nunique().sort_values()
display(user_df)
```

```python
ratings_n = pd.DataFrame(ratings_movies_df.groupby('userId')['rating'].agg('describe'))
display(ratings_n[['count', 'mean']].sort_values(by = ['count', 'mean'], ascending = [True, False]))
```

```python
ratings_genres = ratings_movies_df[ratings_movies_df['Movie year'] == 2018].groupby('genres')['rating'].agg(['mean', 'count'])
display(ratings_genres[ratings_genres['count'] > 10].sort_values(by='mean', ascending=False))
```

```python
orders = pd.read_csv('data/orders.csv', sep = ";")
display(orders)
```

```python
products = pd.read_csv('data/products.csv', sep=';')
display(products)
```

```python
orders_products = orders.merge(
    products,
    left_on='ID товара',
    right_on='Product_ID',
    how = 'left'
)
display(orders_products)
```

```python
orders_products['Overall_price'] = orders_products['Price'] * orders_products['Количество']
display(orders_products)
```
