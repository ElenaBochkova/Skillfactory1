```python
city_bike = city_bike.drop(['start station id', 'end station id'], axis = 1)
city_bike.info()
```

```python
age = 2018 - city_bike['birth year']
city_bike['age'] = age
city_bike = city_bike.drop(['birth year'], axis = 1)
display(city_bike)
```

```python
print(city_bike[city_bike['age'] > 60]['age'].value_counts().sum())
```

```python
city_bike['starttime'] = pd.to_datetime(city_bike['starttime'])
city_bike['stoptime'] = pd.to_datetime(city_bike['stoptime'])
city_bike['delta'] = city_bike['stoptime'] - city_bike['starttime']
display(city_bike)
```

```python
weekday = city_bike['starttime'].dt.day_of_week
mask = (weekday == 5) | (weekday == 6)
city_bike['weekend'] = mask.apply(lambda x: 1 if x else 0)
```

```python
print(city_bike[city_bike['weekend'] == 1]['weekend'].value_counts())
```

```python
print(city_bike['starttime'].dt.hour)
```

```python
def obtain_time_of_day(date_trip):
    an_hour = date_trip.hour
    if an_hour in [0, 1, 2, 3, 4, 5, 6]:
        return 'night'
    elif an_hour in [7, 8, 9, 10, 11, 12]:
        return 'morning'
    elif an_hour in [13, 14, 15, 16, 17, 18]:
        return 'day'
    elif an_hour in [19, 20, 21, 22, 23]:
        return 'evening'
    else:
        return ''
    
city_bike['time of day'] = city_bike['starttime'].apply(obtain_time_of_day)
display(city_bike)
```

```python
day_trip = city_bike[city_bike['time of day'] == 'day']['age'].count()
night_trip = city_bike[city_bike['time of day'] == 'night']['age'].count()
print(round(day_trip / night_trip, 0))
```
